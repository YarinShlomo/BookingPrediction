```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Hotels.csv")
```

```python
# Clean hotel name
newNames = []
for i in range(len(df.Name)):
    name = df.Name[i]
    if '\n' in name:
        name = name.split('\n')[1]
    newNames.append(name)
df['Name'] = newNames
```

```python
# remove duplication of rows
df.drop_duplicates(keep = "first", inplace = True)
```

```python
# Reset df indexing
df.head(50)
```

```python
# Remove rows where stars = 0
df = df[df.Stars != 0]
```

```python
# Remove bad info from rating
df = df[df.Rating != "Read"]
```

```python
df.info()
```

```python
# Remove rows with missing values
df.dropna(axis=0, subset=['Wifi', "Allowing pets", "Closest attraction dist","payment methods"], inplace = True)
```

```python
# replace np.nan -> 1 in Closest transportation col.
df['Closest transportation'].fillna(1, inplace = True)
```

```python
df.reset_index(drop=True, inplace=True)
df.info()
```

```python
# Removing the ',' from numbers with more than 3 digits so we will be able to convert to float. 
df['Price'] = df['Price'].str.replace(',','')
df['Reviews'] = df['Reviews'].str.replace(',','')

# Converting columns to float.
df.astype({'Price':np.float64,'Reviews':np.float64}).dtypes
```

```python
df.to_csv("CleanHotels.csv", index= False)
```
